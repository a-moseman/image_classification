```python
PATH = 'G:/My Drive/data/'
TRAIN_PATH = 'train/'
```

```python
from PIL import Image, ImageOps
import os
import numpy as np
import re
import random
```

```python
IMAGE_HEIGHT = 148
IMAGE_WIDTH = 148
```

```python
filenames = os.listdir(PATH + TRAIN_PATH)
n = 1999
```

# No Preprocessing

```python
img = cv.imread(PATH + TRAIN_PATH + filenames[n])
img = cv.resize(img, (IMAGE_WIDTH, IMAGE_HEIGHT))
data = np.asarray(img).astype(np.uint8) / 255
plt.imshow(data)
```

# Mean Filter

```python
img = cv.imread(PATH + TRAIN_PATH + filenames[n])
img = cv.resize(img, (IMAGE_WIDTH, IMAGE_HEIGHT))
img = cv.blur(img, ksize=(3,3))
data = np.asarray(img).astype(np.uint8) / 255
plt.imshow(data)
```

# NL-Bayes Denoising

```python
img = cv.imread(PATH + TRAIN_PATH + filenames[n])
img = cv.resize(img, (IMAGE_WIDTH, IMAGE_HEIGHT))
img = cv.fastNlMeansDenoisingColored(img)
data = np.asarray(img).astype(np.uint8) / 255
plt.imshow(data)
```

# Canny Edge Detection

```python
img = cv.imread(PATH + TRAIN_PATH + filenames[n])
img = cv.resize(img, (IMAGE_WIDTH, IMAGE_HEIGHT))
img = cv.Canny(img, 100, 200)
data = np.asarray(img).astype(np.uint8) / 255
plt.imshow(data)
```

# NL-Bayes Denoising + Canny Edge Detection

```python
img = cv.imread(PATH + TRAIN_PATH + filenames[n])
img = cv.resize(img, (IMAGE_WIDTH, IMAGE_HEIGHT))
img = cv.fastNlMeansDenoisingColored(img)
img = cv.Canny(img, 100, 200)
data = np.asarray(img).astype(np.uint8) / 255
plt.imshow(data)
```
